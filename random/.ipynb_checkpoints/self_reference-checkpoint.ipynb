We can actually take approximations then as
$$
\pi = 2^N [(\textrm{SQRT}^{N}(-1) - 1)/i]
$$
where $SQRT^{N}(-1)$ represents $\sqrt{ \sqrt{ \sqrt{ \dots \sqrt{-1}}}}$ as $N \rightarrow \infty$. $2^\infty$ and $\textrm{SQRT}^{\infty}$ are effectively fixed points given that $2 * 2^\infty = 2^\infty$ and $\textrm{SQRT} (\textrm{SQRT}^{\infty}(-1)) = \textrm{SQRT}^{\infty}(-1)$. 

```python
import numpy as np
import matplotlib.pyplot as plt
def pi_approx(N):
    val = complex(-1, 0)
    for _ in range(N):
        val = np.sqrt(val)
    approx = 2**N * ((val - 1)/ 1j)
    return approx.real
```

```python
N = np.arange(1, 60, 10, dtype = int)
vals = [pi_approx(n) for n in N]
err = abs(np.pi - np.array(vals))
plt.plot(N, err, "r*")
plt.xlabel("N")
plt.ylabel("|Ï€ - approx(N)|")
```
